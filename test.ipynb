```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']


fig, ax = plt.subplots(2,1, figsize=(10, 8))
x_axis = np.arange(-4, 31)
ax1 = ax[0]
ax2 = ax[1]




hope = np.arange(1,13,dtype=int)
fear = np.arange(1,13,dtype=int)
label = ["(无优劣势)", "(单优势)", "(单劣势)"]
advantage = [np.zeros(6,dtype=int), np.arange(1,7), -np.arange(1,7)]
for i in range(3):
    adv = advantage[i]
    # use matrix to store the all combination [hope,fear,adv] without loop
    hope_matrix, fear_matrix, adv_matrix = np.meshgrid(hope, fear, adv, indexing='ij')
    total = hope_matrix + fear_matrix + adv_matrix
    crit = hope_matrix == fear_matrix
    all_prob = np.prod(total.shape)  # total number of combinations
    # calculate the probability of total > x
    print(f"| {label[i][1:-1]} | 等于 | 成功 | 失败 |")
    print("|---|-------------|---------|------|")

    prob_ = np.zeros(35)  # to store probabilities for totals from -4 to 30
    success_ = np.zeros(35)  # to store success probabilities
    for x in range(-4,31):
        prob = np.sum(total == x) / all_prob *100
        success = np.sum((total >= x) | crit) / all_prob *100
        fail = 100-success
        print(f"| {x:2d} | {prob:11.2f} % | {success:7.2f} % | {fail:6.2f} % |")
        prob_[x+4] = prob
        success_[x+4] = success

    ax1.plot(x_axis, prob_, label='结果概率'+ label[i], marker='o')
    ax2.plot(x_axis, success_, label='成功概率' + label[i], marker='x')

ax1.set_xlabel('DC')
ax1.set_ylabel('等于DC概率 %')
ax2.set_xlabel('DC')
ax2.set_ylabel('成功(>=DC)概率 %')
ax1.legend(loc="center left")
ax2.legend(loc="center right")
ax1.grid()
ax2.grid()
plt.title('骰子点数分布与成功概率')
plt.tight_layout()
plt.show()
```
